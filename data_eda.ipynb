# Exploratory Data Analysis

```python
# Necessary libraires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# allow to display entire df
pd.set_option('display.max_rows', 1000)

plt.style.use('ggplot')
```

```python
# import the dataset
dataset = pd.read_csv('./data/league_data_cleaned_10min.csv')
dataset.head()
```

```python
dataset.drop(columns='Unnamed: 0', axis=1, inplace=True)
dataset.head(10)
```

## I Basic Data exploration

```python
# Steps of EDA:
# 1 - Devide variables into categorical & continuous variables
# 2 - Look if there's any missing values
# 3 - plot barplots of categorical variables
# 4 - plot the distribution of continuous variables
# 5 - look for any outliers in continuous variables
# 6 - log transform the skewed continuous vairables
# 7 - plot the game duration vairable w/ respect to categorial and then continuous variable (log transformed)
```

```python
# Step 1 : Seperate categorical (any feature w/ unique values =< 10) from continuous variables
categorical_variables = [feature for feature in dataset.columns if len(dataset[feature].unique()) <= 10]
continuous_variables = [feature for feature in dataset.columns if len(dataset[feature].unique()) > 10]
```

```python
# Step 2 : Looking for missing values
for feature in dataset.columns:
    print('The feature {} has {} missing values'.format(feature, dataset[feature].isna().sum()))
# There's no missing values in the dataset
```

```python
# setp 3: Represent distribution of categorical variables
for feature in categorical_variables:
    data = dataset.copy()
    values = data[feature].value_counts().values
    index = data[feature].value_counts().index
    plt.bar(index, values)
    plt.title(feature)
    plt.show()
```

```python
# step 4: The distribution of continuous random variables
for feature in continuous_variables:
    sns.distplot(dataset[feature], bins=20)
    plt.show()
```

```python
# skewed data
# - blueWardsPlaced, blueWardsDestroyed, blueAssists, redWardsPlaced, redWardsDestroyed, redKills, redAssists
```

```python
# step 5: Look for any outliers in the continuous variable
for feature in continuous_variables:
    plt.boxplot(dataset[feature])
    plt.title(feature)
    plt.show()
```

```python
# Step 6: Log transform the skewed continuous variables
skewed_variables = ['blueWardsPlaced', 'blueWardsDestroyed', 'blueAssists', 'redWardsPlaced', 'redWardsDestroyed', 'redKills', 'redAssists']
data = dataset.copy()
for feature in skewed_variables:
    if (data[feature]==0).any(): # check if there is any 0 value in column
        pass
    else:
        data[feature] = np.log(data[feature])
                
```

```python
# step 7: plot the GameDuration variable w/ respect to categorical and continuous variable
for feature in continuous_variables:
    plt.scatter(feature, 'GameDuration', data=data)
    plt.title(feature)
    plt.show()
```

```python
for feature in categorical_variables:
    data.groupby(data[feature])['GameDuration'].median().plot(kind='bar')
    plt.show()
```

```python
# Comparing the cat/continuous features to gameDuration don't yield a lot of information yet. We might want to do some 
# feature engineering to extract some more relevant information : 
# - compute the absolute value of the total difference of kills in the game
# - compute the abs value of the total difference of gold in the game
# - compute the abso value of difference of dragons / herald / tower  killed / destroyed
# - we can also compute the kda difference of each teams
```

## II. Data exploration of engineered features

Since our original features seem to have some poor impact on the variable we are trying to predict (GameDuration) in the first part of **Feature engineering** I build up some new features that capture the difference of performance of the two teams at 10 minutes of the game. Let's perform some basic data exploration on those new features

```python
df_fe = pd.read_csv('data/data_basic_fe.csv')
```

```python
df_fe.head()
```

we replicate the same steps performed on the orignal dataset:
* 1 - Devide variables into categorical & continuous variables
* 2 - Look if there's any missing values
* 3 - plot barplots of categorical variables
* 4 - plot the distribution of continuous variables
* 5 - look for any outliers in continuous variables
* 6 - log transform the skewed continuous vairables
* 7 - plot the game duration vairable w/ respect to categorial and then continuous variable (log transformed)


```python
# Step 1 : Seperate categorical (any feature w/ unique values =< 10) from continuous variables
categorical_variables = [feature for feature in df_fe.columns if len(df_fe[feature].unique()) <= 10]
continuous_variables = [feature for feature in df_fe.columns if len(df_fe[feature].unique()) > 10]
```

```python
# Step 2: checking for nan values -> None we can skip
```

```python
# setp 3: Represent distribution of categorical variables
for feature in categorical_variables:
    data = df_fe.copy()
    values = data[feature].value_counts().values
    index = data[feature].value_counts().index
    sns.barplot(index, values, palette='Accent_r')
    plt.title(feature)
    plt.show()
```

```python
# step 4: The distribution of continuous random variables
for feature in continuous_variables:
    sns.distplot(df_fe[feature], bins=20, color='purple')
    plt.show()
```

Pretty much are the continuous variables are skewed (except GameDuration) we will handle this further in the **Feature Engineering** part.

```python
# step 5: Look for any outliers in the continuous variable
for feature in continuous_variables:
    plt.boxplot(df_fe[feature])
    plt.title(feature)
    plt.show()
```

```python
# Step 6: Log transform the skewed continuous variables that don't have 0 values 
skewed_variables = [x for x in continuous_variables if x != 'GameDuration']
data = df_fe.copy()
for feature in skewed_variables:
    if (data[feature]==0).any(): # check if there is any 0 value in column
        pass
    else:
        data[feature] = np.log(data[feature])
```

```python
# step 7: plot the GameDuration variable w/ respect to categorical and continuous variable
for feature in continuous_variables:
    plt.scatter(feature, 'GameDuration', data=data, color='yellow')
    plt.title(feature)
    plt.show()
```

```python
for feature in categorical_variables:
    data.groupby(data[feature])['GameDuration'].median().plot(kind='bar', color='gray')
    plt.show()
```

It seems that we gain some better insights out of those new features:
* For instance we can see that games where the the total difference of Kills / Assists / Deaths is larger tend to end up earlier. 
* Similarly games where we observe an early game lead (which means that the team which has done the first blood is also the one with the most gold at the end of the first 10 min of game) from one team tend to last a be lower on average.
* Finally the difference between towers destroyed and Elite Monsters killed seem to have a heavy impact on the game duration.

It's now time to get back to some **Feature Engineering** to perform some skewness correction / scale transformation / outlier detection on our new features.
